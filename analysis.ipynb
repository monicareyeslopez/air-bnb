```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load in datasets
listings = pd.read_csv('listings.csv')
reviews = pd.read_csv('reviews.csv')
neighborhoods = pd.read_csv('neighborhoods.csv')
calendar = pd.read_csv('calendar.csv.gz',compression='gzip')
```

### <a id='toc1_1_1_'></a>[listings info](#toc0_)

```python
listings.info()
```

```python
# unnecessary columns to drop:
unnecessary_columns = ['id', 'name','host_id', 'host_name','neighbourhood_group','latitude','longitude','license']
listings_clean = listings.copy()
listings_clean = listings_clean.drop(columns=unnecessary_columns)

#changing name of neighbourhood column to neighborhood becuase we are american
listings_clean = listings_clean.rename(columns={'neighbourhood': 'neighborhood'})
listings_clean.info()
```
